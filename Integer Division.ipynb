```python
%load_ext Cython
```

```python
%%cython --annotate

cdef unsigned char a, b, c, d
a = 7
b = 3
c = a // b
d = a % b

print(c, d)
```

```python
%%cython --annotate

cdef unsigned char a, b, c, d
a = 7
b = 3
c, d = divmod(a, b)

print(c, d)
```

```python
%%cython --annotate

cdef unsigned char a, b, c, d
a = 7
b = 3
c = a / b
d = a - c * b

print(c, d)
```
